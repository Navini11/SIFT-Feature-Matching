SIFT feature matching

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread(r'store\store\store_shelf.png', cv.IMREAD_REDUCED_COLOR_2)
assert image is not None
template1 = cv.imread(r'store\store\oat_crunch.jpg', cv.IMREAD_REDUCED_COLOR_2)
assert template1 is not None
template2 = cv.imread(r'store\store\corn_chex.jpg', cv.IMREAD_REDUCED_COLOR_2)
assert template1 is not None

sift = cv.SIFT_create()
kp1, des1 = sift.detectAndCompute(image, None)
kp2, des2 = sift.detectAndCompute(template1, None)
kp3, des3 = sift.detectAndCompute(template2, None)
```

```python
print(f"Number of keypoints in first image: {len(kp1)}")
print(f"Number of keypoints in second image: {len(kp2)}")
print(f"Number of keypoints in second image: {len(kp3)}")
```

```python
FLAAN_INDEX_KDTREE = 1
index_params = dict(algorithm=FLAAN_INDEX_KDTREE, trees=5)
search_params = dict(checks=50)
flann = cv.FlannBasedMatcher(index_params, search_params)
matches1 = flann.knnMatch(des1,des2,k=2)
matches2 = flann.knnMatch(des1,des2,k=2)

#Need to draw only good matches, so create a mask
matchesMask1 = [[0,0] for i in range(len(matches1))]
#ratio test as per Lowe's paper
for i,(m,n) in enumerate(matches1):
    if m.distance < 0.7*n.distance:
        matchesMask1[i]=[1,0]

dra_params1 = dict(matchColor = (0,255,0),
                  singlePointColor = (255,0,0),
                  matchesMask = matchesMask1,
                  flags = cv.DrawMatchesFlags_DEFAULT) 

img3 = cv.drawMatchesKnn(image, kp1, template1, kp2, matches1, None, **dra_params1)
plt.imshow(cv.cvtColor(img3, cv.COLOR_BGR2RGB))
plt.show()

#Need to draw only good matches, so create a mask
matchesMask2 = [[0,0] for i in range(len(matches2))]
#ratio test as per Lowe's paper
for i,(m,n) in enumerate(matches2):
    if m.distance < 0.7*n.distance:
        matchesMask2[i]=[1,0]

dra_params2 = dict(matchColor = (0,255,0),
                  singlePointColor = (255,0,0),
                  matchesMask = matchesMask2,
                  flags = cv.DrawMatchesFlags_DEFAULT) 

img4 = cv.drawMatchesKnn(image, kp1, template2, kp3, matches2, None, **dra_params2)
plt.imshow(cv.cvtColor(img4, cv.COLOR_BGR2RGB))
plt.show()
```
